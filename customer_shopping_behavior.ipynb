```python
df["Review Rating"]=df.groupby("Category")["Review Rating"].transform(lambda x:x.fillna(x.median()))  # here we are finding out the median review rating within each category and will replace it that with the missing value
```

```python
df.isnull().sum()
```

```python
# here we are converting all the colunms in the snake case i.e. every letter of the columns name will be small and spaces between them will be replaced by the underscore.
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(" ","_")
df.columns
df=df.rename(columns={"purchase_amount_(usd)":"purchase_amount"})
```

**Feature Engineering**

```python
# create a columns age_group

labels=["Young Adult","Adult","Middle-Aged","Senior"]
df["age_group"]=pd.qcut(df["age"],q=4,labels=labels)   #qcut will splits the ages into four equal size group based on the data distribution and assign the label with define
```

```python
df[["age","age_group"]].head(10)
```

```python
#creating a columns purchase_frequency_days

frequency_mapping={
    "Fortnightly": 14,
    "Weekly": 7,
    "Monthly": 30,
    "Quarterly": 90,
    "Bi-Weekly": 14,
    "Annually": 365,
    "Every 3 Months": 90
}
df["purchase_frequency_days"]=df['frequency_of_purchases'].map(frequency_mapping)  # here first we have created a dictionary called frequency_mapping in which we have assign the number for the corresponding to the text value of the "frequency_of_purchases" column and then we are using the map function to convert the text value into numeric value.
```

```python
df[["purchase_frequency_days","frequency_of_purchases"]].head(10)  # we can see both columns the text columns , the numeric columns
```

```python
df[["discount_applied","promo_code_used"]].head(10)
```

```python
(df["discount_applied"]==df["promo_code_used"]).all()   # this means both the columns i.e. discount_applied and promo_code_used caries exactly the same information. so we can remove promo code .
```

```python
df=df.drop("promo_code_used", axis=1)
```
